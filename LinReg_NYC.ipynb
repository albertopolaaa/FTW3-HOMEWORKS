```python
del data['Unnamed: 0']
del data['SALE DATE']
del data ['EASE-MENT']
```

```python
data['SALE PRICE'] = pd.to_numeric(data['SALE PRICE'], errors='coerce') #convert dtype to numeric
data['GROSS SQUARE FEET'] = pd.to_numeric(data['GROSS SQUARE FEET'], errors='coerce') #convert dtype to numeric
data['LAND SQUARE FEET'] = pd.to_numeric(data['LAND SQUARE FEET'], errors='coerce') #convert dtype to numeric
```

```python
data['SALE PRICE'] = data['SALE PRICE'].fillna(0) #null = 0
data['GROSS SQUARE FEET'] = data['GROSS SQUARE FEET'].fillna(0)
data['LAND SQUARE FEET'] = data['LAND SQUARE FEET'].fillna(0)
```

```python
features = ['BBL', 'BUILDING CLASS CATEGORY','TAX CLASS AT PRESENT','BUILDING CLASS AT PRESENT', 'TAX CLASS AT TIME OF SALE','BUILDING CLASS AT TIME OF SALE','GROSS SQUARE FEET', 'LAND SQUARE FEET','COMMERCIAL UNITS','YEAR BUILT', 'NEIGHBORHOOD', 'ADDRESS','TOTAL UNITS']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = data[features]
y = data['SALE PRICE']
```

```python
scaler = StandardScaler()
scaler.fit(X)

transformed_data = scaler.transform(X)
transformed_data_df = pd.DataFrame(transformed_data, columns=X.columns)
```

```python
sns.set(rc={'figure.figsize':(20,10)})

from sklearn.model_selection import train_test_split # split dataset

from sklearn.linear_model import LogisticRegression # linear model for classification

from sklearn.metrics import confusion_matrix

from sklearn.metrics import mean_squared_error as mse
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
# initialize model
log_reg = LogisticRegression()

# fit model
log_reg.fit(X_train, y_train)

# metrics (accuracy)
acc = log_reg.score(X_test, y_test)


# predictions
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr
est = rfr(n_estimators=100, random_state=0).fit(X_train, y_train)
a = est.predict(X_test)
```
